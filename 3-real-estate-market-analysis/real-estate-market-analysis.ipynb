–ü–æ—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫–∏–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å–æ–¥–µ—Ä–∂–∏—Ç –∫–æ–ª–æ–Ω–∫–∞.

–ù–∞–∑–≤–∞–Ω–∏–π –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ –æ–≥—Ä–æ–º–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ, –ø—Ä–∏—á–µ–º –µ—Å—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è, –Ω–æ —Å —Ä–∞–∑–Ω—ã–º –Ω–∞–ø–∏—Å–∞–Ω–∏–µ–º... –†–µ—à–µ–Ω–æ,—á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –ø—É—Ç–∞–Ω–∏—Ü—ã –æ—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–∞—Å.–ø—É–Ω–∫—Ç–æ–≤ –±–µ–∑ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º—É —Ç–∏–ø—É. 

````python
type_locality_name=['–ø–æ—Å—ë–ª–æ–∫', '–≥–æ—Ä–æ–¥—Å–∫–æ–π', '–¥–µ—Ä–µ–≤–Ω—è',
                    '–≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ', '—Ç–∏–ø–∞', '–ø–æ—Å–µ–ª–æ–∫', '—Å–∞–¥–æ–≤–æ–µ', '—Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ', 
                    '—Å–µ–ª–æ', '—Å–∞–¥–æ–≤–æ–¥—á–µ—Å–∫–æ–µ', '–Ω–µ–∫–æ–º–º–µ—Ä—á–µ—Å–∫–æ–µ', '—Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ', '–∫–æ—Ç—Ç–µ–¥–∂–Ω—ã–π','–ø—Ä–∏', '–∂–µ–ª–µ–∑–Ω–æ–¥–æ—Ä–æ–∂–Ω–æ–π' ] #–¥–µ–ª–∞–µ–º —Å–ø–∏—Å–æ–∫ –¥–ª—è —Å–ª–æ–≤, –æ—Ç –∫–æ—Ç–æ—Ä—ã—Ö –º—ã —Ö–æ—Ç–∏–º –∏–∑–±–∞–≤–∏—Ç—Å—è.

for name in type_locality_name: # —Å–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª, –∫–æ—Ç–æ—Ä—ã–º –ø—Ä–æ–π–¥–µ–º—Å—è –ø–æ –≤—Å–µ–º—É —Å–ø–∏—Å–∫—É
    data['locality_name'] = data['locality_name'].str.replace(name, '')
data['locality_name'] = data['locality_name'].str.lstrip() # –º–µ—Ç–æ–¥–æ–º strip() –∏–∑–±–∞–≤–∏–º—Å—è –æ—Ç –ø—Ä–æ–±–µ–ª–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥–ª–∏ –æ–±—Ä–∞–∑–æ–≤–∞—Ç—å—Å—è –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è —Å–ª–æ–≤.

print(data['locality_name'].unique()) # –≤—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç

````

–¢–µ–ø–µ—Ä—å –¥–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –∑–∞–ø–æ–ª–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–æ–ª–æ–Ω–∫–∞—Ö airports_nearest, city_centers_nearest, parks_around3000, parks_nearest ,ponds_around3000 ,ponds_nearest , —Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∫–æ–ª–æ–Ω–∫–µ locality_name  –≤—ã—Å—á–∏—Ç–∞–µ–º –º–µ–¥–∏–∞–Ω—É –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ –∏ –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏. –¢–∞–º –≥–¥–µ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –º–µ–¥–∏–∞–Ω—É –≤ –≥—Ä—É–ø–ø–µ, –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–æ –≤—Å–µ–º—É –Ω–∞–±–æ—Ä—É –¥–∞–Ω–Ω—ã—Ö.

````python
groups = data.groupby('locality_name') #—Å–æ—Ö—Ä–∞–Ω–∏–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—è–º –Ω–∞—Å.–ø—É–Ω–∫—Ç–æ–≤
for name, group in groups: #–ø—Ä–æ–π–¥–µ–º—Å—è —Ü–∏–∫–ª–æ–º –ø–æ –≥—Ä—É–ø–ø–∞–º
    airports_rows = group.query('airports_nearest > 0') #—Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ query –Ω–∞–π–¥–µ–º –≥—Ä—É–ø–ø—ã, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –¥–∞–Ω–Ω—ã–µ –¥–ª—è –≤—ã—á–µ—Å–ª–µ–Ω–∏—è –º–µ–¥–∏–∞–Ω—ã –µ—Å—Ç—å –∏ —Å–æ—Ö—Ä–∞–Ω–∏–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π  
    if len(airports_rows) > 0: #–ø—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å–ª–∏ –µ—Å—Ç—å –≤ —ç—Ç–æ–π –≥—Ä—É–ø–ø–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é
        group_median = airports_rows['airports_nearest'].median() #–≤—ã—Å—á–∏—Ç—ã–≤–∞–µ–º –º–µ–¥–∏–∞–Ω—É –¥–ª—è –≥—Ä—É–ø–ø—ã 
        data.loc[(data['locality_name'] == name) & (data['airports_nearest'].isna()), 'airports_nearest'] = group_median #–∑–∞–ø–æ–ª–Ω—è–µ–º –º–µ–¥–∏–∞–Ω–æ–π –≥—Ä—É–ø–ø—ã
        

data.loc[data['airports_nearest'].isna(), 'airports_nearest'] = data['airports_nearest'].median() # –∑–∞–ø–æ–ª–Ω—è–µ–º –æ–±—â–µ–π –º–µ–¥–∏–∞–Ω–æ–π –≤—Å–µ –æ—Å—Ç–∞–≤—à–∏–µ—Å—è –ø—Ä–æ–ø—É—Å–∫–∏ , –ø—Ä–∏–º–µ–Ω—è—è –º–µ—Ç–æ–¥ –ª–æ–≥–∏—á–µ—Å–∫–æ–π –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏

data['airports_nearest'].isna().sum() #–ø—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞—à–µ–≥–æ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–æ–ø—É—Å–∫–æ–≤
````

–¢–æ–∂–µ —Å–∞–º–æ–µ –¥–µ–ª–∞–µ–º –∏ —Å –æ—Å—Ç–∞–ª—å–Ω—ã–º–∏ –∫–æ–ª–æ–Ω–∫–∞–º–∏

````python

def floor_group(floor): #—Ñ—É–Ω–∫—Ü–∏—é –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ø—Ä–∏–º–µ–Ω—è–µ–º –∫ —Å—Ç–æ–ª–±—Ü—É 'floor'
     
    if floor <= 1 :
        return '–ø–µ—Ä–≤—ã–π'
    if floor > 1 and floor <= 32:
        return '–¥—Ä—É–≥–æ–π'
    return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'

print(floor_group(1)) #–ø—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞–±–æ—Ç—É 
print(floor_group(5))
print(floor_group(33))
print()
data['floor_group'] = data['floor'].apply(floor_group) #—Å–æ–∑–¥–∞–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –≥—Ä—É–ø–ø–∞–º–∏ 


print(data['floor_group'].head())
````
